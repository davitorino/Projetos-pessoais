# Parte 2 - Implementação das Métricas de Distância

```python

df_teste = df.set_index('id_cliente')

# função de distância
def distancia_euclidiana(cliente1, cliente2):
    variaveis = ['idade', 'valor_compra', 'num_compras', 'tempo_cliente_dias']
    v1 = cliente1[variaveis].astype(float)
    v2 = cliente2[variaveis].astype(float)
    dif = v1 - v2
    return np.sqrt((dif ** 2).sum())

# selecionar clientes
cliente_a = df_teste.loc['CLI001']
cliente_b = df_teste.loc['CLI002']
cliente_c = df_teste.loc['CLI003']

dist_ab = distancia_euclidiana(cliente_a, cliente_b)
dist_ac = distancia_euclidiana(cliente_a, cliente_c)

print(f"Distância Euclidiana entre CLI001 e CLI002: {dist_ab:.2f}")
print(f"Distância Euclidiana entre CLI001 e CLI003: {dist_ac:.2f}")
```

# Parte 3 - Pipe Line e Limpeza de Dados


```python
import pandas as pd
import numpy as np

def pipeline_limpeza() -> tuple[pd.DataFrame, dict]:
    df = pd.read_csv('/content/sample_data/Dataset_DataMartePlus.csv')
    relatorio = {}

    # Etapa 1: Remoção de Duplicatas
    duplicatas_iniciais = df.duplicated(subset='cpf').sum()
    df = df.sort_values('data_cadastro', ascending=False)
    df = df.drop_duplicates(subset='cpf', keep='first')
    relatorio['duplicatas_removidas'] = int(duplicatas_iniciais)

    # Etapa 2: Correção de Tipos
    df['valor_compra'] = (
        df['valor_compra'].astype(str).str.replace(',', '.', regex=False).str.strip()
    ).astype(float)

    # Etapa 3: Tratamento de Valores Faltantes
    mediana_geral = df['idade'].median()
    idades_novas = []
    for _, linha in df.iterrows():
        if pd.isna(linha['idade']):
            grupo = df[(df['genero'] == linha['genero']) & (df['estado'] == linha['estado'])]
            mediana_grupo = grupo['idade'].median()
            if not np.isnan(mediana_grupo):
                idades_novas.append(mediana_grupo)
            else:
                idades_novas.append(mediana_geral)
        else:
            idades_novas.append(linha['idade'])
    df['idade'] = idades_novas
    relatorio['idades_preenchidas'] = int(df['idade'].isna().sum())

    # Etapa 4: Detecção e Tratamento de Outliers
    Q1 = df['valor_compra'].quantile(0.25)
    Q3 = df['valor_compra'].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    mediana_valor = df['valor_compra'].median()
    cond_outlier = (df['valor_compra'] < limite_inferior) | (df['valor_compra'] > limite_superior)
    qtd_outliers = int(cond_outlier.sum())
    print(f"Foram encontrados {qtd_outliers} outliers em 'valor_compra'.")
    df.loc[cond_outlier, 'valor_compra'] = mediana_valor
    relatorio['outliers_corrigidos'] = qtd_outliers

    # Etapa 5: Validação Final
    relatorio['nulos_restantes'] = int(df.isna().sum().sum())
    relatorio['duplicatas_finais'] = int(df.duplicated(subset='cpf').sum())
    relatorio['shape_final'] = tuple(df.shape)

    return df, relatorio

# CHAMADA DA FUNÇÃO E PRINTS
print("Antes da limpeza:")
df_raw = pd.read_csv('/content/sample_data/Dataset_DataMartePlus.csv')
print("Shape:", df_raw.shape)
print("Nulos por coluna:\n", df_raw.isna().sum())
print("Duplicatas:", df_raw.duplicated(subset='cpf').sum())

df_limpo, rel = pipeline_limpeza()

print("\nDepois da limpeza:")
print("Shape:", df_limpo.shape)
print("Nulos por coluna:\n", df_limpo.isna().sum())
print("Duplicatas:", df_limpo.duplicated(subset='cpf').sum())
print("\nRelatório:", rel)
```

# Parte 4: Análise de Simularidade

```python
# Parte 4: Análise de Similaridade
# 4.1 Encontrar Cliente Mais Similar

# Usar o dataframe limpo
df_analise = df_limpo.copy()

# Variáveis numéricas para normalização Min-Max
vars_num = ['idade', 'valor_compra', 'num_compras', 'tempo_cliente_dias']

# Normalizar as variáveis (Min-Max para [0, 1])
for col in vars_num:
    min_val = df_analise[col].min()
    max_val = df_analise[col].max()
    if max_val > min_val:
        df_analise[f'{col}_norm'] = (df_analise[col] - min_val) / (max_val - min_val)
    else:
        df_analise[f'{col}_norm'] = 0  # Caso de variação zero

# Definir índice por id_cliente
df_analise = df_analise.set_index('id_cliente')

# Variáveis normalizadas
vars_norm = [f'{v}_norm' for v in vars_num]

# Funções de distância usando variáveis normalizadas
def distancia_euclidiana_norm(cliente1, cliente2):
    v1 = cliente1[vars_norm].values
    v2 = cliente2[vars_norm].values
    dif = v1 - v2
    return np.sqrt((dif ** 2).sum())

def distancia_manhattan_norm(cliente1, cliente2):
    v1 = cliente1[vars_norm].values
    v2 = cliente2[vars_norm].values
    return np.abs(v1 - v2).sum()

# Cliente de referência (exemplo: CLI001)
cliente_ref_id = 'CLI001'
if cliente_ref_id not in df_analise.index:
    print(f"Cliente {cliente_ref_id} não encontrado. Usando o primeiro disponível.")
    cliente_ref_id = df_analise.index[0]

cliente_ref = df_analise.loc[cliente_ref_id]

# Calcular distâncias para todos os outros clientes
dists_eucl = []
dists_manh = []

for idx, row in df_analise.iterrows():
    if idx == cliente_ref_id:
        continue
    dist_e = distancia_euclidiana_norm(cliente_ref, row)
    dist_m = distancia_manhattan_norm(cliente_ref, row)
    dists_eucl.append((idx, dist_e))
    dists_manh.append((idx, dist_m))

# Ordenar e selecionar top 3 (menores distâncias)
top3_eucl = sorted(dists_eucl, key=lambda x: x[1])[:3]
top3_manh = sorted(dists_manh, key=lambda x: x[1])[:3]

# Exibir resultados
print(f"\nCliente de referência: {cliente_ref_id}")
print("\nTop 3 clientes mais similares por Distância Euclidiana:")
for i, (id_cl, dist) in enumerate(top3_eucl, 1):
    print(f"{i}. {id_cl}: {dist:.4f}")

print("\nTop 3 clientes mais similares por Distância Manhattan:")
for i, (id_cl, dist) in enumerate(top3_manh, 1):
    print(f"{i}. {id_cl}: {dist:.4f}")


```

```python
# 4.2 Implementação da Função (Versão Corrigida)
def encontrar_similares(cliente_referencia, df_clientes, metrica='euclidiana', top_n=3):
    import pandas as pd
    import numpy as np

    # Copia o DataFrame
    df = df_clientes.copy()

    # Variáveis numéricas
    vars_num = ['idade', 'valor_compra', 'num_compras', 'tempo_cliente_dias']

    # Verifica se as variáveis existem
    for v in vars_num:
        if v not in df.columns:
            raise ValueError(f"Coluna '{v}' não encontrada no DataFrame.")

    # Normalização Min-Max
    for col in vars_num:
        min_val = df[col].min()
        max_val = df[col].max()
        if max_val > min_val:
            df[f'{col}_norm'] = (df[col] - min_val) / (max_val - min_val)
        else:
            df[f'{col}_norm'] = 0

    # Variáveis normalizadas
    vars_norm = [f'{v}_norm' for v in vars_num]

    # Define índice por id_cliente se não estiver definido
    if 'id_cliente' in df.columns:
        df = df.set_index('id_cliente')

    # Cliente de referência
    if cliente_referencia not in df.index:
        print(f"Aviso: Cliente '{cliente_referencia}' não encontrado. Usando o primeiro cliente disponível.")
        cliente_referencia = df.index[0]

    cliente_ref = df.loc[cliente_referencia]

    # Funções de distância
    def dist_euclidiana(c1, c2):
        v1 = c1[vars_norm].values
        v2 = c2[vars_norm].values
        return np.sqrt(np.sum((v1 - v2) ** 2))

    def dist_manhattan(c1, c2):
        v1 = c1[vars_norm].values
        v2 = c2[vars_norm].values
        return np.sum(np.abs(v1 - v2))

    # Calcula distâncias
    distancias = []
    dist_func = dist_euclidiana if metrica == 'euclidiana' else dist_manhattan

    for idx, row in df.iterrows():
        if idx == cliente_referencia:
            continue
        dist = dist_func(cliente_ref, row)
        distancias.append((idx, dist))

    # Ordena e seleciona top_n (menores distâncias)
    distancias.sort(key=lambda x: x[1])
    top_clientes = distancias[:top_n]

    # Cria DataFrame de retorno
    resultado = pd.DataFrame(top_clientes, columns=['Cliente', 'Distancia'])
    resultado.index = range(1, len(resultado) + 1)
    resultado.index.name = 'Rank'

    return resultado, cliente_referencia

# 4.3 Comparação das Métricas
cliente_ref = 'CLI001'  # Ajuste se souber o ID correto

# Obtém top 3 para cada métrica usando df_limpo
top_eucl, ref_usado_eucl = encontrar_similares(cliente_ref, df_limpo, metrica='euclidiana', top_n=3)
top_manh, ref_usado_manh = encontrar_similares(cliente_ref, df_limpo, metrica='manhattan', top_n=3)

# Usa o cliente referência efetivo
cliente_ref_final = ref_usado_eucl  # Assume o mesmo para ambas

# Cria tabela de comparação
comparison_data = {
    'Rank': [1, 2, 3],
    'Euclidiana': top_eucl['Cliente'].values,
    'Dist_Eucl': top_eucl['Distancia'].round(2).values,
    'Manhattan': top_manh['Cliente'].values,
    'Dist_Manh': top_manh['Distancia'].round(2).values
}

comparison_df = pd.DataFrame(comparison_data)
print(f"\nComparação para o cliente de referência: {cliente_ref_final}")
print(comparison_df.to_string(index=False))


```
